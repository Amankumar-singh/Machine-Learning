```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head() # By using head command we can access first 5 row
```

```python
sns.barplot(x = 'sex', y ='total_bill' , data = tips)
# now this is mean of all bill
```

```python
# we can also add estimator = np.std
sns.barplot(x = 'sex' , y = 'total_bill' , data = tips , estimator= np.std)
```

```python
sns.countplot(x = 'sex' , data = tips)  #  in countplot we can't change agrrestio parameter means y function
```

```python
sns.countplot(x= "time" , data = tips)
```

```python
sns.boxplot(x="day", y="total_bill", data = tips)
```

```python
# hue is a parameter
sns.boxplot(x="day" , y="total_bill" , data = tips , hue = 'sex')
```

```python
sns.violinplot(x='day' , y = 'total_bill' , data = tips , hue = "sex" )
```

```python
sns.violinplot(x='day' , y = 'total_bill' , data = tips , hue = "sex" , split = True) # split will just combine 
```

```python
sns.distplot(tips['total_bill'])
```

```python
# if we don't want to print histogram then we can use kdeplot
sns.kdeplot(tips['total_bill'])
```

```python
# joint plot is a another type of plot which also shows relation :- It is between two graph means two varaible
sns.jointplot(x='total_bill' , y='tip' , data = tips)  # show here we can see relation between totall bill and tip

# here if we want to visualize in different way then there is option kind and it's parameter
```

```python
# pairplot ,It plot pairwise rlationship among all the feature of data frame

sns.pairplot(tips) # we don't have to give x, y because it will print pairwise all combination in data set
```

```python
# we can use hue parameter here 
sns.pairplot(tips, hue='sex') # It will just plot male and female diffrent plot 
```

```python
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr)  # if you have light color mrans high co-relations , if you have dark color means low co-relations
```

```python
#if we want to display number then we can use  annot parameter 
sns.heatmap(tips_corr , annot = True)
```
