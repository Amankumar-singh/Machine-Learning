```python
import matplotlib.pyplot as plt
import numpy as np   
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1)
plt.show()
```

```python
# we can see available theme

theme = plt.style.available
print(theme)
```

```python
# Now here we are using one of theme
plt.style.use("seaborn")
plt.plot(x,y1)
plt.show()
```

```python
# WE can use color in graph also
plt.plot(x,y1, color="green")
plt.plot(x,y2, color="red")
plt.show()
```

```python
# we can also use lable , title  and labble markig on graph and his style
plt.plot(x,y1, color='green',label="Apple",marker='o')
plt.plot(x,y2, color="red",label="Kiwi",linestyle="dashed",marker='*')
plt.xlabel("Time")
plt.xlabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend(loc="lower right")
plt.show()
```

```python
prices = np.array([1,2,3,4])**4
print(prices)
plt.plot(prices) # If we want to plot a graph by single array then in that case the given array will
                 # work as Y and about x then it will take automatically array of his index
                 # example in this case Y = [1,16,81,256] and X = [0,1,2,3]
plt.show()  
```

# **Scatter Plots**


```python
# we can also define size of figure 
plt.figure(figsize=(4,4)) # it will make the size according to parameter passes to it
plt.scatter(x,y1,color='red',marker='o',label="Apple")
plt.scatter(x,y2,color='green',marker='^',label='Kiwi')
plt.xlabel("Time")
plt.ylabel('Price')
plt.title("Scatter Plots - Prices of fruits over Time")
plt.legendz()
plt.show()
```

# **Bar-graphs**

```python
plt.bar([0,1,2],[10,20,15])
plt.show()
```

```python
plt.bar([0,1,2],[10,20,15])
plt.bar([0,1,2],[20,10,12])
plt.show()
#these both are overlapig not look good 
```

```python
# now hwere we will use commom x cordinate but slighlty we will make different
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current",tick_label=["Gold","Platinum","Silver"])
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year")
plt.title("Metal Price Comparision")
plt.title("Metal price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")

plt.legend()
plt.show
```
